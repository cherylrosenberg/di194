```python
# ============================================
# RUN THIS CELL FIRST - All imports for today
# ============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Display settings
plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 20)

# Set default figure size
plt.rcParams['figure.figsize'] = (10, 6)

print("All libraries loaded! Ready for visualization.")
```

```python
days_web = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [150, 200, 180, 220, 300, 250, 180]

print("Visitor data:")
for day, visitor_count in zip(days_web, visitors):
    print(f"  {day}: {visitor_count}")
```

```python
# Customized Line Plot - Professional Quality
plt.figure(figsize=(10, 6))

plt.plot(days_web, visitors,
         marker='o',           # Add points at each value
         linewidth=2,          # Thicker line
         markersize=8,         # Bigger points
         color='steelblue')    # Professional color

plt.title('Daily Website Visitors', fontsize=16, fontweight='bold')
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Number of Visitors', fontsize=12)
plt.grid(True, alpha=0.3)  # Subtle gridlines
plt.ylim(140, 310)  # Set Y-axis range to emphasize variation

plt.show()

print("Same data, much more professional!")
```

```python
categories = ['Electronics', 'Clothing', 'Food', 'Books']
product_sales = [45000, 32000, 28000, 15000]

plt.bar(categories, product_sales, color='teal', edgecolor='black')
plt.title('Sales by Product Category', fontsize=14, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.show()
```
