### Your Task

* Download and import the Data Science Job Salary dataset.
* Normalize the ‘salary’ column using Min-Max normalization which scales all salary values between 0 and 1.
* Implement dimensionality reduction like Principal Component Analysis (PCA) or t-SNE to reduce the number of features (columns) in the dataset.
* Group the dataset by the ‘experience_level’ column and calculate the average and median salary for each experience level (e.g., Junior, Mid-level, Senior).

```python
import pandas as pd
```

```python
salaries = pd.read_csv('datascience_salaries.csv')
salaries
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Set up the figure with two subplots to compare before/after
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# 2. Visualize the original distribution
sns.histplot(salaries['salary'], kde=True, ax=ax1, color='skyblue')
ax1.set_title('Original Salary Distribution')
ax1.set_xlabel('Salary (Raw Value)')

# 3. Box Plot (shows quartiles and potential outliers)
sns.boxplot(x=salaries['salary'], ax=ax2, color='lightgreen')
ax2.set_title('Salary Box Plot')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Initialize the scaler
scaler = MinMaxScaler()

# Reshape the data (Scikit-Learn expects a 2D array) and transform
# Note: we use [[ ]] to keep it as a DataFrame/2D array
salaries['salary_normalized'] = scaler.fit_transform(salaries[['salary']])

print(salaries[['salary', 'salary_normalized']].head())
```

```python
# 2. Setup the figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# 3. Plot Original
sns.histplot(salaries['salary'], kde=True, ax=ax1, color='skyblue')
ax1.set_title('Original Salary (Skewed)')

# 4. Plot Normalized
sns.histplot(salaries['salary_normalized'], kde=True, ax=ax2, color='salmon')
ax2.set_title('Min-Max Normalized (0 to 1)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Select categorical columns to encode
cat_cols = ['job_title', 'job_type', 'experience_level', 'location']

# One-hot encode categorical features
encoder = OneHotEncoder(sparse_output=False)
encoded_features = encoder.fit_transform(salaries[cat_cols])

# Combine normalized salary with encoded features
X = np.hstack([salaries[['salary_normalized']].values, encoded_features])
```

```python
# Reduce to 2 principal components
pca = PCA(n_components=2)
pca_results = pca.fit_transform(X)

salaries['pca_1'] = pca_results[:, 0]
salaries['pca_2'] = pca_results[:, 1]
```

```python
# Reduce to 2 dimensions for visualization
# Note: perplexity should be smaller than your number of samples
tsne = TSNE(n_components=2, perplexity=2, random_state=42)
tsne_results = tsne.fit_transform(X)

salaries['tsne_1'] = tsne_results[:, 0]
salaries['tsne_2'] = tsne_results[:, 1]
```

```python
# 3. Visualize and Compare
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot PCA
sns.scatterplot(x=pca_results[:,0], y=pca_results[:,1], hue=salaries['job_title'], ax=ax1)
ax1.set_title('PCA: Preserves Global Variance')

# Plot t-SNE
sns.scatterplot(x=tsne_results[:,0], y=tsne_results[:,1], hue=salaries['job_title'], ax=ax2)
ax2.set_title('t-SNE: Preserves Local Clusters')

plt.show()
```

```python
# Group by experience level and calculate mean and median
experience_stats = salaries.groupby('experience_level')['salary'].agg(['mean', 'median'])

# Sorting the results so they are easier to read (e.g., Junior to Senior)
# This assumes your experience levels follow a logical order
print(experience_stats)
```

```python
# Compare raw vs normalized across experience levels
summary = salaries.groupby('experience_level')[['salary', 'salary_normalized']].agg(['mean', 'median'])
print(summary)
```
