1. Data Preparation:
* Download and explore the US Superstore data.
* Perform basic data cleaning and preprocessing.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
path = r'c:\Users\chery\Documents\di194\Week3\Day5\US Superstore data.xls'
superstore_data = pd.read_excel(path)

# Check the first few rows and data types
superstore_data.info()
```

```python
superstore_data.head(10)
```

```python
# Checking for null values
superstore_data.isnull().sum()
```

```python
# Convert postal code to a string to avoid calculations on the value
superstore_data['Postal Code'] = superstore_data['Postal Code'].astype(str)

# Verify the change
superstore_data['Postal Code'].dtype
```

```python
# Check for duplicate rows
duplicate_count = superstore_data.duplicated().sum()
print(f"Number of duplicated rows: {duplicate_count}")
```

```python
#Calculate shipping time
superstore_data['Ship Time'] = (superstore_data['Ship Date'] - superstore_data['Order Date']).dt.days
```

```python
# Create a profit margin column
superstore_data['Profit Margin %'] = (superstore_data['Profit'] / superstore_data['Sales']) * 100

superstore_data[['Sales', 'Profit', 'Profit Margin %']].head()
```

```python
# Box plot analysis to check for outliers in the sales and profit columns 

fig, axes = plt.subplots(1,2, figsize=(15,6))

# Boxplot for sales 
sns.boxplot(ax=axes[0], y=superstore_data['Sales'], color='skyblue')
axes[0].set_title('Box Plot of Sales')

# Boxplot for Profit
sns.boxplot(ax=axes[1], y=superstore_data['Profit'], color='salmon')
axes[1].set_title('Box Plot for Profit')

plt.tight_layout()
plt.show()
```

2. Data Visualization with Matplotlib:
* Create an interactive line chart to show sales trends over the years.
* Build an interactive map to visualize sales distribution by country.

```python
# Prepare the data
superstore_data['Year'] = superstore_data['Order Date'].dt.year
yearly_sales = superstore_data.groupby('Year')['Sales'].sum().reset_index()

# Create the plot
plt.figure(figsize=(10,6))
plt.plot(yearly_sales['Year'], yearly_sales['Sales'], marker='o', linestyle='-', color='b', linewidth=2)

# Customization
plt.title('Total Sales Trends Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(yearly_sales['Year'])
plt.grid(True, linestyle='--', alpha=0.7)

# Add labels for clarity
for i, txt in enumerate(yearly_sales['Sales']):
    plt.annotate(f'${txt:,.0f}', (yearly_sales['Year'][i], yearly_sales['Sales'][i]), textcoords='offset points', xytext=(1,10), ha='center')

plt.show()
```

```python
# Prepare the data
state_dist = superstore_data.groupby('State')['Sales'].sum().sort_values(ascending=False)

# Create the plot 
plt.figure(figsize=(12,8))
state_dist.plot(kind='bar', color='teal')

# Customization
plt.title('Sales Distribution by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

3. Data Visualization with Seaborn:
* Use Seaborn to generate a bar chart showing top 10 products by sales.
* Create a scatter plot to analyze the relationship between profit and discount.

```python
# Aggregate sales by Product Name
top_10_products = superstore_data.groupby('Product Name')['Sales'].sum().reset_index()
top_10_products = top_10_products.sort_values(by='Sales', ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(12,8))
sns.barplot(data=top_10_products, x='Sales', y='Product Name', hue='Product Name', legend=False, palette='viridis')

# Add labels and Title
plt.title('Top 10 Products by Total Sales', fontsize=15)
plt.xlabel('Total Sales ($)', fontsize=12)
plt.ylabel('Product Name', fontsize=12)

# Add data labels
for container in plt.gca().containers:
    plt.gca().bar_label(container, fmt='$%.0f', padding=3)

plt.show()
```

```python
# Create the Scatter Plot
plt.figure(figsize=(12,7))
sns.scatterplot(data=superstore_data, x='Discount', y='Profit', hue='Category', alpha=0.6)

# Add a horizontal line at 0 to clearly see losses
plt.axhline(0, color='red', linestyle='--')

# Customization
plt.title('Relationship between Discount and Profit', fontsize=15)
plt.xlabel('Discount (0.0 to 0.8)', fontsize=12)
plt.ylabel('Profit / Loss ($)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)

plt.show()
```
