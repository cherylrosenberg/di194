In the new era of AI, data analysis in the world of fraud prevention has been transformed. The advancements in capability have enabled financial institutions to not only identify fraud that's already occurred, but catch fraud before it happens. In 2025 / early 2026 Mastercard analyzed transactional data such as amount, merchant type, location and fruquency of purchases. They also analyzed behavioural patterns, such as how a user holds their phone, typing speed, and angle of the device during the transaction. They analyzed the unique ID of the phone, the individual's historical patterns, and real-time cyber threat intelligence. 
Using all of this intelligent data, along with AI, Mastercard was able to shift from if/then rules of fraud flagging to Graph Neural Networks, which were used to find "hidden" fraud rings that were undetectable to just pure human intervention. The models are also able to predict how new types of scams might look before they happen. The system was able to analyze and interfere in milliseconds, prevening fraud before it happened. Without all of this data, as well as the use of AI to analyze as much as 160 billion transactions annually, hidden patterns wouldn't have been found.
