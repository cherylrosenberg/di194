```python
data = {
    "city": ["Paris", "Paris", "Paris", "Paris",
             "London", "London", "London", "London",
             "Rome", "Rome", "Rome", "Rome"],
    "year": [2001, 2008, 2009, 2010,
             2001, 2006, 2011, 2015,
             2001, 2006, 2009, 2012],
    "pop": [2.148, 2.211, 2.234, 2.244,
            7.322, 7.657, 8.174, 8.615,
           2.547, 2.627, 2.734, 2.627]
}
```

```python
census = pd.DataFrame(data)
```

```python
# Save to the Colab session storage
census.to_excel('census_data.xlsx', index=False)
```

```python
# 'records' format is usually the most common (list of dictionaries)
census.to_json('census_data.json', orient='records', indent=4)
```
